```python
u, d, v = np.linalg.svd(a)
print('u')
print(u)
print('d',d)
print('v',v)
```

```python
d_pse = np.zeros((a.shape[0], a.shape[1])).T

d_pse
```

```python
print("Valores a reemplazar en d_pse")
print(d_pse[:d.shape[0], :d.shape[0]])
```

```python
print("Valores a que pondremos en d_pse")
print(np.linalg.inv(np.diag(d)))
```

```python
print("d_pse")
d_pse[:d.shape[0], :d.shape[0]] = np.linalg.inv(np.diag(d))

d_pse
```

```python
a_pse = v.T.dot(d_pse).dot(u.T)

a_pse
```

```python
a_pse_calc = np.linalg.pinv(a)

a_pse.dot(a)
```

## Usando la pseudoinversa para resolver sistemas sobredeterminados

```python
x = np.linspace(-5, 5, 1000)

y_1 = -1*x+4
y_2 = 2*x+5
y_3 = -3*x+6

plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)

plt.xlim(-2,2.5)
plt.ylim(-6, 6)
plt.axhline(y=0, color='gray')
plt.axvline(x=0, color='gray')

np.linalg.pinv([y_1, y_2, y_3])
```

```python
matriz = np.array([[4,1], [-2,1], [3,1]])

matriz
```

```python
matriz_pse = np.linalg.pinv(matriz)
matriz_pse
```

```python
b = np.array([[3],[5],[1]])

resultado = matriz_pse.dot(b)
print(resultado)
```

```python
from glob import iglob

caras = []

for path in iglob('./imagenes/*/*.pgm'):
  im = imageio.imread(path)
  cara = pd.Series(im.flatten(), name=path)
  caras.append(cara)

fig, axes = plt.subplots(5,10, figsize=(15,8), subplot_kw={'xticks': [], 'yticks':[]}, gridspec_kw=dict(hspace = 0.01, wspace=0.01))

caras = pd.DataFrame(caras)

for i, ax in enumerate(axes.flat):
  ax.imshow(caras.iloc[i].values.reshape(112,92), cmap='gray')
```

```python
from sklearn.decomposition import PCA

caras_pca = PCA(n_components=0.5)
```

```python
caras_pca.fit(caras)
```

```python
filas = 3
columnas = caras_pca.n_components_

fig, axes = plt.subplots(filas,columnas, figsize=(12,6), subplot_kw={'xticks': [], 'yticks':[]}, gridspec_kw=dict(hspace = 0.01, wspace=0.01))

caras = pd.DataFrame(caras)

for i, ax in enumerate(axes.flat):
  ax.imshow(caras_pca.components_[i].reshape(112,92), cmap='gray')
```

```python
caras_pca.n_components_
```

```python
componentes = caras_pca.transform(caras)
proyeccion = caras_pca.inverse_transform(componentes)
```

```python
fig, axes = plt.subplots(5,10, figsize=(15,8), subplot_kw={'xticks': [], 'yticks':[]}, gridspec_kw=dict(hspace = 0.01, wspace=0.01))

for i, ax in enumerate(axes.flat):
  ax.imshow(proyeccion[i].reshape(112,92), cmap='gray')
```
